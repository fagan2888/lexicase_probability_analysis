```python
# examing the effect of population size and test case size on probability of selection.

# probabilitiy of a case coming first at least once

import numpy as np
import itertools as it
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

%matplotlib inline
plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['xtick.labelsize'] = 14

N = np.arange(2,1000,10)
T = np.arange(2,2000,10)

N,T = np.meshgrid(N,T)

Pcf = 1 - ((T-1)/T)**N

# for (n,t) in it.product(N,T):
#     X.append(n)
#     Y.append(t)
#     Pcf.append(1 - ((t-1)/t)**n)
    
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(N, T, Pcf, cmap = cm.coolwarm,linewidth=0,antialiased=False,alpha=0.9)
ax.set_xlabel('\nN',size=16)
ax.set_ylabel('\nT',size=16)
ax.set_zlabel('Pr',size=16)
ax.set_zlim(0.0, 1.0)
ax.set_xlim(0,1000)
ax.set_ylim(0,2000)
ax.xaxis._axinfo['label']['space_factor'] = 2.8
ax.yaxis._axinfo['label']['space_factor'] = 1.0
ax.zaxis._axinfo['label']['space_factor'] = 2.8

# ax.set_zticks(np.arange(0,1.1,0.1))
# ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))
ax.view_init(30, 130) 
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)


fig.tight_layout()
fig.savefig('../../../Papers/epsilon_lexicase_journal/figs/prob_case_first.pdf',figsize=(6,4))
fig.savefig('../../../Papers/epsilon_lexicase_journal/figs/prob_case_first.png')
plt.show()
```
